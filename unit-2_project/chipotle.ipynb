### Part 3: Calculate the average price of an order.

Hint: Examine the data to see if the `quantity` column is relevant to this calculation.

Hint: Think carefully about the simplest way to do this!

```python
# Examine each column in data
for row in data:
    print('order_id:', row[0], type(row[0]))
    print('quantity:', row[1], type(row[1]))
    print('item_name:', row[2], type(row[2]))
    print('choice_description:', row[3], type(row[3]))
    print('item_price:', row[4], type(row[4]))
    print('\n')
```

```python
# Convert the order_id column from string to integer
# Convert the quantity column from string to integer
# Convert the item_price column from string to float, and remove '$''
for row in data:
    row[0] = int(row[0])
    row[1] = int(row[1])
    row[4] = float(row[4].split('$')[1])
    
    print('order_id:', row[0], type(row[0]))
    print('quantity:', row[1], type(row[1]))
    print('item_price:', row[4], type(row[4]))
    print('\n')
```

```python
# Calculate the total price of each order
dict_order_totalprice = {}
i = 1
order_totalprice = 0

for row in data:
    if row[0] == i:
        order_totalprice += row[4]
        dict_order_totalprice[row[0]] = order_totalprice
    else:
        order_totalprice = row[4]
        dict_order_totalprice[row[0]] = order_totalprice
        i += 1

dict_order_totalprice
```

```python
# Calculate the average price of an order, round off to 2 decimals
order_averageprice = float(sum(dict_order_totalprice.values()))/len(dict_order_totalprice)
print('$' + str(round(order_averageprice, 2)))
```

### Part 4: Create a list (or set) named `unique_sodas` containing all of unique sodas and soft drinks that Chipotle sells.

Note: Just look for `'Canned Soda'` and `'Canned Soft Drink'`, and ignore other drinks like `'Izze'`.

```python
# Look for 'Canned Soda' and 'Canned Soft Drink' in data
all_orders_sodas = []

for row in data:
    if ('Canned Soda' in row[2]) or ('Canned Soft Drink' in row[2]):
        if row[1:] not in all_orders_sodas:
            all_orders_sodas.append(row[1:])

all_orders_sodas.sort()
all_orders_sodas
```

```python
# Check for duplicates; order_id does not matter
unique_orders_sodas = []

for row in all_orders_sodas:
    if row[0] == 1:
        unique_orders_sodas.append(row)
    else:
        row[3] = row[3] / row[0]
        row[0] = 1
        if row not in unique_orders_sodas:
            unique_orders_sodas.append(row)           

unique_orders_sodas.sort()
unique_orders_sodas
```

```python
# Store all unique sodas and soft drinks in unique_sodas
unique_sodas = []

for row in unique_orders_sodas:
    unique_sodas.append(row[2])

# Remove duplicates    
unique_sodas = set(unique_sodas)
unique_sodas
```

### Part 5: Calculate the average number of toppings per burrito.

Note: Let's ignore the `quantity` column to simplify this task.

Hint: Think carefully about the easiest way to count the number of toppings!


```python
# Look for orders with burrito in data
all_orders_burritos = []

for row in data:
    if 'Burrito' in row[2]:
        if row not in all_orders_burritos:
            all_orders_burritos.append(row[2:4])

all_orders_burritos.sort()
all_orders_burritos
```

```python
# Calculate the total number of toppings per burrito
all_toppings_burritos = []

for row in all_orders_burritos:
    all_toppings_burritos.append((row[1].count(',')) + 1)

print(all_toppings_burritos)
```

```python
# Calculate the average number of toppings per burrito
average_toppings_burritos = float(sum(all_toppings_burritos))/len(all_toppings_burritos)
print(round(average_toppings_burritos))
```

### Part 6: Create a dictionary. Let the keys represent chip orders and the values represent the total number of orders.

Expected output: `{'Chips and Roasted Chili-Corn Salsa': 18, ... }`

Note: Please take the `quantity` column into account!

Optional: Learn how to use `.defaultdict()` to simplify your code.

```python
# Look for orders with chip in data; quantity matters
all_chips_quantity = []

for row in data:
    if 'Chip' in row[2]:
        all_chips_quantity.append(tuple(row[1:3][::-1]))

all_chips_quantity.sort()
all_chips_quantity
```

```python
# Create a dictionary; quantity matters
from collections import defaultdict

dict_chips_quantity = defaultdict(int)

for key, value in all_chips_quantity:
    dict_chips_quantity[key] += value

dict_chips_quantity
```
